```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sorted_data skinaqua.csv')
df.head(10)
```

```python
df['Recommend'] = df['Recommend'].map({'recommend': 1, 'not recommend': 0})
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Pilih kolom numerik saja untuk korelasi
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numerical_cols].corr()

# Visualisasi correlation matrix dengan heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data untuk produk Skinaqua UV Moisture Gel
filtered_df = df[df['Jenis_Produk'] == 'Skinaqua UV Moisture Gel']

# Pastikan kolom 'rating' tersedia di filtered_df
if 'rating' not in filtered_df.columns:
    print("Kolom 'rating' tidak ditemukan dalam dataset.")
else:
    # Plot distribusi rating
    plt.figure(figsize=(8, 6))
    sns.countplot(data=filtered_df, x='rating', palette='viridis')
    plt.title('Distribusi Rating untuk Skinaqua UV Moisture Gel', fontsize=14)
    plt.xlabel('Rating', fontsize=12)
    plt.ylabel('Jumlah', fontsize=12)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Menghitung jumlah pengguna untuk setiap Jenis_Produk
product_counts = df['Jenis_Produk'].value_counts()

# Warna khusus: pink, ungu, dan biru
colors = ['#FFC0CB', '#9370DB', '#87CEEB']  # Pink, ungu, biru

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    product_counts,
    labels=product_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors
)
plt.title('Distribusi Pengguna Berdasarkan Jenis Produk', fontsize=14)
plt.axis('equal')  # Membuat pie chart berbentuk lingkaran sempurna
plt.tight_layout()
plt.show()
```
