```python
print(df['age'].unique())
```

```python
# Define the mapping for age ranges to 4 categories
age_mapping = {
    '18 and Under': 'remaja',     # Teenagers
    '19 - 24': 'dewasa muda',    # Young adults
    '25 - 29': 'dewasa muda',
    '30 - 34': 'dewasa matang',  # Mature adults
    '35 - 39': 'dewasa matang',
    '40 - 44': 'dewasa matang',
    '45 and Above': 'paruh baya'        # Older adults
}

# Apply the mapping to the 'age' column
df['age_category'] = df['age'].map(age_mapping)

# Print unique values of the new 'age_category' column
print("Unique values in 'age_category':")
print(df['age_category'].unique())
```

```python
print(df['profile_description'].unique())
```

```python
# Rename the column 'profile_description' to 'skintype'
df.rename(columns={'profile_description': 'skintype'}, inplace=True)

# Check the unique values before filtering
print("Unique values in 'skintype' before filtering:")
print(df['skintype'].unique())

# Remove rows with specific unwanted values
unwanted_values = [',19 - 24', ',25 - 29', ',18 and Under']
df = df[~df['skintype'].isin(unwanted_values)]  # Keep rows not in the unwanted_values list

# Check the unique values after filtering
print("\nUnique values in 'skintype' after filtering:")
print(df['skintype'].unique())
```

```python
print(df['usage_period'].unique())
print(df['rating_count'].unique())
print(df['Jenis Produk'].unique())
```

```python
# Mapping periode penggunaan ke dalam tiga kategori dalam bahasa Indonesia
usage_period_mapping_id = {
    'Less than 1 week': 'Jangka Pendek',
    '1 week - 1 month': 'Jangka Pendek',
    '1 month - 3 months': 'Jangka Pendek',
    '3 months - 6 months': 'Jangka Menengah',
    '6 months - 1 year': 'Jangka Menengah',
    'More than 1 year': 'Jangka Panjang'
}

# Terapkan mapping pada kolom 'usage_period'
df['periode_penggunaan'] = df['usage_period'].map(usage_period_mapping_id)

# Cetak nilai unik dari kolom baru
print("Nilai unik dalam 'periode_penggunaan':")
print(df['periode_penggunaan'].unique())
```

```python
#rename column
df.rename(columns={'rating_count': 'rating'}, inplace=True)
print(df['rating'].unique())
df.rename(columns={'Jenis Produk': 'Jenis_Produk'}, inplace=True)
print(df['Jenis_Produk'].unique())
df.rename(columns={'value': 'trend'}, inplace=True)
print(df['trend'].unique())
```

```python
df.drop(columns=['usage_period'], inplace=True)
df.drop(columns=['age'], inplace=True)
df.head(10)
```

```python
# Mapping nilai dalam kolom 'recommend' menjadi kategori 'recommend' dan 'not recommend'
df['Recommend'] = df['recommend'].apply(
    lambda x: 'recommend' if 'recommends this product!' in str(x).lower() else 'not recommend'
)

# Tampilkan nilai unik dalam kolom baru
print(df['Recommend'].unique())
drop = ['recommend']
df.drop(columns=drop, inplace=True)
df.head(10)
```

```python
# Memisahkan kolom 'skintype' menjadi beberapa bagian berdasarkan koma
df[['skintype', 'skintone', 'undertone']] = df['skintype'].str.split(',', expand=True)

# Membersihkan nilai yang dihasilkan (menghapus spasi tambahan)
df['skintype'] = df['skintype'].str.strip()
df['skintone'] = df['skintone'].str.strip()
df['undertone'] = df['undertone'].str.strip()

# Tampilkan beberapa baris pertama untuk memverifikasi hasil
print(df[['skintype', 'skintone', 'undertone']].head())
```

```python
df.head(10)
```

```python
# Sorting berdasarkan kolom 'Jenis_Produk'
df_sorted = df.sort_values(by='Jenis_Produk', ascending=True)

# Tampilkan data yang sudah diurutkan
print(df_sorted.head(10))  # Menampilkan beberapa baris pertama
```

```python
# Simpan df_sorted ke dalam file CSV
df_sorted.to_csv('sorted_data.csv', index=False)

# Simpan df_sorted ke dalam file Excel
df_sorted.to_excel('sorted_data.xlsx', index=False)

print("Data telah disimpan ke 'sorted_data.csv' dan 'sorted_data.xlsx'")
```
